# Trying out the Azure OpenAI embedding endpoint using the params: azure_endpoint, api_key, and api_version

```python
import os
from dotenv import load_dotenv

load_dotenv("/home/stepanogil/ChatBotDev/jg-api-bot-backend/.env")

from openai import AzureOpenAI


azure_endpoint = os.getenv("BCFG_AZURE_OPENAI_ENDPOINT") 
# print(azure_endpoint)
api_key=os.getenv("BCFG_AZURE_OPENAI_API_KEY")  
# print(api_key)
api_version=os.getenv("BCFG_AZURE_OPENAI_API_VERSION")
# print(api_version)

client = AzureOpenAI(
  azure_endpoint = azure_endpoint,   
  api_key = api_key, 
  api_version = api_version
)
```

### It works fine as evidendenced by the generated embeddings

```python
response = client.embeddings.create(
    input = "in the time of chimpanzees I was a monkey",
    model= "text-embedding-ada-002"
)

print(response.model_dump_json(indent=2))
```

## Using LangChain AzureOpenAIEmbedding and AzureSearch classes

```python
from langchain.embeddings import AzureOpenAIEmbeddings
from langchain.vectorstores.azuresearch import AzureSearch
```

### used the same params for AzureOpenAIEmbeddings

```python

embeddings = AzureOpenAIEmbeddings(model="text-embedding-ada-002", 
                                   chunk_size=400, 
                                   api_key=api_key,
                                   azure_endpoint=azure_endpoint, 
                                   api_version=api_version,)
print(embeddings)
```
